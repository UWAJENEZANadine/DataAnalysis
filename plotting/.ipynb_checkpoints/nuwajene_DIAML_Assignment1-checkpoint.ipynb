```python
#question 3

#variable declaration principal(p), annual interest rate(r), number of times (n)

p = 100
r = 0.05 #which is equal to 5%
n = 1

FirstYear = p * (1 + (r/n))**(n*1)
SecondYear = p * (1 + (r/n))**(n*2)
ThirdYear = p * (1 + (r/n))**(n*3)
FourthYear = p * (1 + (r/n))**(n*4)
FiveYear = p * (1 + (r/n))**(n*5)

print(f"Amount of money you will have after First year: ${round(FirstYear)}")
print(f"Amount of money you will have after Second year: ${round(SecondYear)}")
print(f"Amount of money you will have after Third year: ${round(ThirdYear)}")
print(f"amount of money you will have after Fourth year: ${round(FourthYear)}")
print(f"amount of money you will have after Five year: ${round(FiveYear)}")
```

```python
#question 4

#variable declaration principal(p) and monthly interest rate

p = 20000
r = 0.01  #Montly interest rate is given as 1%

FirstYear = p * r * (1 + r)**12 / ((1 + r)**12 -1)
SecondYear = p * r * (1 + r)**24 / ((1 + r)**24 -1)
ThirdYear = p * r * (1 + r)**36 / ((1 + r)**36 -1)

print(f"Monthly pay off in one year: ${FirstYear:.2f}")
print(f"Monthly pay off in second year: ${SecondYear:.2f}")
print(f"Monthly pay off in Third year: ${ThirdYear:.2f}")
```

```python
#question 5


#import matloplotlib library for helps to plot our data
import matplotlib.pyplot as plt

#variable declaration business investment(I), customer expected per day(C), growth rate per day(G) and profit per customer(p)
I = 100000
C = 100
G = 0.01 #growth rate 1%
P = 10


#Calculation of days to break even by using while loop

m = 0
cumulative_profit = 0
array =[]
while cumulative_profit < I:
    m += 1
    profit_per_day = C * P
    cumulative_profit += profit_per_day
    C *= (1 + G)
    array.append(cumulative_profit)
    
#out put of the number of days break even

print("The number of days break even it will take", m)

#The plotting of number of cumulated profits per day by showing initial investment and mark brakeven day
plt.plot(array)
plt.axhline(y=I, color ='g')
plt.text(0,I + 5000, "Initial investment", color= 'r')
plt.text(m - 1, 6000, f"Breakeven Day:{m}", color='b')
plt.plot(m-1, I, 'o', color='black') #to makr the break even day
plt.xlabel('Days')
plt.ylabel('cumulative profit')
# plt.grid(True)
```

```python
#question 6

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#variable to load data
filename = 'ebola_download.xls'
#define dataframe for reading a value

df = pd.read_excel(filename, engine= 'xlrd')

#Interpolation of data using linear interpolation
x = np.arange(len(df)) #create an array of integer from o to the length of dataframe df
y_cases = df['Cases']
x_death = df['Death']
x_new = np.linspace(x[0], x[-1], num=len(x), endpoint=True) #This new array of x-coardinates is used to plot the interpolated data, which allow us to estimate the number of cases and deaths due to ebola at any point of time
new_y_cases = np.interp(x_new, x, y_cases)
new_y_death = np.interp(x_new, x, x_death)


#Thresholds to be marked with circle
thresholds = [100, 500, 1000, 2000, 5000]

fig, axes = plt.subplots() #create new plot of interpolated data
axes.plot(x_new, new_y_cases, label = 'Cases')# x-new array contains the x-coardinates for the plot, while new_y_cases contains the interpolatated values for the number of cases.
axes.plot(x_new, new_y_death, label = 'Death')# x_new array contains the x-coordinates for the plot, while new_y_death contains the interpolated values for the number of deaths, and label is set to death which specipy the labe to plot



#the thresholds to marked exceed with the circle is defined
for t in thresholds:
    index = np.where(new_y_cases > t)[0]
    if len(index) > 0:
        index = index[0]
    date = df['Date'][index]
    axes.plot(index, t, 'o', color = 'green')
    axes.annotate(date.strftime('%y-%m-%d'), (index+10, t+50)) #This annotate the plot with the date when a threshold was exceed. (index+10, threshold+50) specifies the location of the annotation on the plot.
    

#set the plots
axes.set_xlabel('Days')
axes.set_ylabel('Number of cases/death')

axes.legend()# this legend provides visual guide to the meaning of the different lines and markers on the plot
plt.show() #to display a plot with show()
```

```python
#question 7

import pandas as pd
import numpy as np

#load the data
filename = 'ebola_download.xls'

df = pd.read_excel(filename, engine = 'xlrd')
df [['Date', 'Cases', 'Death', 'Diff', 'Noofdays']]
df.columns = ['Date', 'Cases', 'Death', 'Diff', 'Noofdays']

#date column to datetime changed to
df['Date'] = pd.to_datetime(df['Date'])

#date column as index
df.set_index('Date', inplace=True)

#daily growth rate of cases and deaths is calculated as 
df['cases_gr'] = df['Cases'].pct_change()  * 100
df['deaths_gr'] = df['Death'].pct_change() *100

#daily average growth rate of cases and deaths is calculated as
avg_gr_cases = df['cases_gr'].mean()
avg_gr_deaths = df['deaths_gr'].mean()

#out put of average dayily growth rate for death and cases
print ('average growth rate per day of cases',avg_gr_cases)
print ('average growth rate per day of deathes', avg_gr_deaths)

```

```python
#question 10


import pandas as pd

filename1 = 'SPY.csv'
filename2 = 'TLT.csv'

#starting and ending date declaration
start_date = '2014-01-01'
end_date = '2015-08-31'

#getting closing prices
spyfile = pd.read_csv(filename1, usecols =['Date', 'Adj Close'], index_col = 'Date', parse_dates= True)
tltfile = pd.read_csv(filename2, usecols =['Date', 'Adj Close'], index_col = 'Date', parse_dates = True)

#calculation of daily returns
spy_return = spyfile.pct_change()
tlt_return = tltfile.pct_change()

#calculation of Average, minimum and maximum

avg_spy_return = spy_return.mean() * 100
min_spy_return = spy_return.min() * 100
max_spy_return = spy_return.max() * 100

avg_tlt_return = tlt_return.mean() * 100
min_tlt_return = tlt_return.min() * 100
max_tlt_return = tlt_return.max() * 100


#the out puts
print(f"Average daily return for SPY:{round(avg_spy_return,2)}")
print("Minimum daily return for SPY:{:.2f}%",min_spy_return)
print("Maximum daily return for SPY:{:.2f}%",max_spy_return)


print("Average daily return for TLT:{:.2f}%",avg_tlt_return)
print("Minimum daily return for TLT:{:.2f}%",min_tlt_return)
print("Maximum daily return for TLT:{:.2f}%",max_tlt_return)



```
